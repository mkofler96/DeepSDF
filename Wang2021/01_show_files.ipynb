```python
import scipy.io
import matplotlib.pyplot as plt
import random
mat = scipy.io.loadmat('Geometries.mat')
geoms = mat["Geometries"]
mat = scipy.io.loadmat('Properties.mat')
properties = mat["Properties"]
```

```python
fig, axs = plt.subplots(5,5, figsize=(20,20))
for ax in axs.flatten():
    i = random.randint(0,geoms.shape[2])
    ax.imshow(geoms[:,:,i])
    ax.axis("off")
    ax.set_title(i)
```
